```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
r=np.sum(s,axis=1)
c=np.sum(s,axis=0)
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n=s[i:i+3,j:j+3]
```

```python
a=np.all(r == 45)
b=np.all(c == 45)
d=n.sum()
```

```python
print((a and b and d) == 45)
```

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],  # Student 1
    [19, 92, 88],  # Student 2
    [17, 76, 95],  # Student 3
    [18, 65, 70],  # Student 4
    [20, 90, 85]   # Student 5
])
```

```python
#get shape of matrix
data.shape
```

```python
#find average age
print(np.mean(data[:,0]))
```

```python
#extract math marks
data[:,1]
```

```python
#find highest science marrks
print(np.max(data[:]))
```

```python
#get detail of student who scored more than 90 in Maths
data[data[:,1]>90]
```

```python
#increase maths marks by 5
data[:,1]+=5
data
```

```python
#how many student younger than 19
len(data[data[:,0]<19])
```

```python
#column wise average
np.mean(data[:,1:],axis=0)
```

```python
# Details of students who scored at least 80 marks in both subjects
data[(data[:,1:] > 79).all(axis=1)]
```

```python
#replace all marks of science <75 with 0
data[:, 2][data[:, 2] < 75] = 0
data
```
