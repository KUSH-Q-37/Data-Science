MERGE 2 DATAFRAMES

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right')
```

on -- common column name

how -- how you merge data
  
  inner -- default or intersection
  
  outer -- union
  
  left -- left table full and both table common
  
  right -- right table full and both table common

  if any value is not present NaN is place of it.

Concatination of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2],axis= 1)
```

.concate - use to concate 2 dataframes 

axis=0 is default or if we write axis =0 --> common column and concate in order

if we write axis =1 --> common row and concate in order

Joining 2 DataFrames

```python
#first DataFrame
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df1.join(df2,how='inner')
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```

```python
df1.join(df2)
```
