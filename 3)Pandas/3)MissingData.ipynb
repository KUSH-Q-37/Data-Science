```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

Finding Missing Data

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

.isna() -- Check Null Values bool


.sum() -- Give Count Column-wise Null Values


.any() -- Check Null Values Column-wise bool

Removing Missing Data

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

Removing is Done Row-wise.


.dropna()-remove full row if null value present.


.dropna(thresh = n )- remove row which have atleast n non- Null values.

Filling the missing data

```python
df
```

```python
df.fillna('n')
```

```python
df.fillna(2)
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```

.fillna(n): we can fill null values with any integer,float,string value

.fillna(value=values) - we have to create a dictionary and then the null values fill column-wise.

.fillna(df.mean()) - we have more mode,median,mean.
